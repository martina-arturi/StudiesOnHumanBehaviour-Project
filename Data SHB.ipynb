### Locationeventpertime

```python
import pandas as pd

location_pd = pd.read_csv('locationeventpertime.csv')
location_pd
```

This table is composed by:

- Timestamp: show year(4), month(2), day(2), hour(2), minute(2), second(2), decimals(3) (Eliminate second and decimals)
- Accuracy: accuracy in meters. (Eliminate)
- Bearing: numeric value (Eliminate)
- Lucene: null (Eliminate)
- Point: gps value (latitude, longitude, altitude)
- Provider: network / gps (Eliminate)
- Speed: numeric speed (Eliminate)

```python
location_new = location_pd
location_new.pop('accuracy')
location_new.pop('bearing')
location_new.pop('lucene')
location_new.pop('provider')
location_new.pop('speed')
location_new
```

```python
location_new.to_csv('location_new.csv')
```

It doesn't make sense to have seconds and decimals in my timestamp, so I will remove them and group by not only the timestamp but also the point column, in this way the information of the changes of places won't get lost but if I stayed in the same place for hours it won't show again and again 

```python
import csv
with open('location.csv', 'w') as new:
    writer = csv.writer(new)
    with open('location_new.csv') as loc_file:
        reader = csv.reader(loc_file, delimiter = ',')

        for el in reader:
                if len(el[1]) == 17:
                    el[1] = el[1][:-5]
                writer.writerow(el)
    new.close()
```

Once I cleaned the timestamp I can import the new file in pandas to group it by timestamp

```python
df = pd.read_csv('location.csv')
df.pop('Unnamed: 0')
new = df.groupby(['timestamp', 'point'])
new.sum().to_csv('finallocation.csv')
```

```python
loc = pd.read_csv('finallocation.csv')
loc
```

### Applicationevent

```python
import pandas as pd

app_pd = pd.read_csv('applicationevent.csv')
app_pd
```

This table is similar to the one before, as we have:
- Timestamp: show year(4), month(2), day(2), hour(2), minute(2), second(2), decimals(3)
- Applicationname: name of the application

It doesn't make sense to have seconds and decimals in my timestamp, so I will remove them and group by not only the timestamp but also the applicationname column


```python
import csv
with open('app.csv', 'w') as new:
    writer = csv.writer(new)
    with open('applicationevent.csv') as app_file:
        reader = csv.reader(app_file, delimiter = ',')

        for el in reader:
                if len(el[0]) == 17:
                    el[0] = el[0][:-5]
                writer.writerow(el)
    new.close()
```

```python
df = pd.read_csv('app.csv')
new = df.groupby(['timestamp', 'applicationname'])
new.sum().to_csv('finalapp.csv')

app = pd.read_csv('finalapp.csv')
app
```

It doesn't make sense to have seconds and decimals in my timestamp, so I will remove them and group by not only the timestamp but also the package column

```python
import csv
with open('notification.csv', 'w') as new:
    writer = csv.writer(new)
    with open('notification_new.csv') as not_file:
        reader = csv.reader(not_file, delimiter = ',')

        for el in reader:
                if len(el[1]) == 17:
                    el[1] = el[1][:-5]
                writer.writerow(el)
    new.close()
```
