### Locationeventpertime

```python
import pandas as pd

location_pd = pd.read_csv('locationeventpertime.csv')
location_pd
```

This table is composed by:

- Timestamp: show year(4), month(2), day(2), hour(2), minute(2), second(2), decimals(3) (Eliminate second and decimals)
- Accuracy: accuracy in meters. (Eliminate)
- Bearing: numeric value (Eliminate)
- Lucene: null (Eliminate)
- Point: gps value (latitude, longitude, altitude)
- Provider: network / gps (Eliminate)
- Speed: numeric speed (Eliminate)

```python
location_new = location_pd
location_new.pop('accuracy')
location_new.pop('bearing')
location_new.pop('lucene')
location_new.pop('provider')
location_new.pop('speed')
location_new
```

```python
location_new['timestamp'] = pd.to_datetime(location_new['timestamp'],format='%Y%m%d%H%M%S%f').round('min')
location_new
```

```python
import csv
with open('coord.csv', 'w') as f:
    my_writer = csv.writer(f)

    for el in location_new['point']:
        a = el.split(',')
        b = [el.replace('{', '').replace('}', '').replace('latitude:', '').replace('longitude:', '').replace('altitude:', '')]
        my_writer.writerow(b)
        
    f.close()
```

```python
df = pd.read_csv('coord.csv')
df = df.rename(columns = {'46.05606,11.12064,273.0': 'coordinates'})
df = df["coordinates"].str.split(",", n = 2, expand = True)
df = df.rename(columns = {0: 'lat', 1: 'long', 2: 'alt'})
df
```

### Applicationevent
